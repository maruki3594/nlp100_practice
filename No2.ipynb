```python
col1_path = './col1.txt'
col2_path = './col2.txt'
l = [s.split() for s in l]
print(l[:][1])

with open(col1_path, mode='w') as f:
    f.writelines('\n'.join([i[0] for i in l]))
    
with open(col2_path, mode='w') as f:
    f.writelines('\n'.join([i[1] for i in l]))
```

```python
with open(col1_path) as f:
    col1 = f.readlines()
col1 = [s.replace('\n', '') for s in col1]
print(col1)
    
with open(col2_path) as f:
    col2 = f.readlines()
col2 = [s.replace('\n', '') for s in col2]
print(col2)

with open('merge_col1_col2.txt', mode='w') as f:
    f.writelines('\n'.join([f'{col1[i]}\t{col2[i]}' for i in range(len(col1))]))
    f.write('\n')
```

```python
import math
n = int(input(prompt='n='))

with open('merge_col1_col2.txt') as f:
    l = f.readlines()
    
for i in range(n):
    with open('merge_col1_col2_' + str(i) + '.txt', mode='w') as f:
        f.writelines(''.join([l[i * math.ceil(len(l) / n) + j] for j in range(math.ceil(len(l) / n)) if i * math.ceil(len(l) / n) + j < len(l)]))
```

```python
import collections

with open('popular-names.txt') as f:
    l = f.readlines()

l = [s.replace('\t', ' ').replace('\n', '') for s in l]
l = [s.split() for s in l]
result = list(set([s[0] for s in l]))
print(result)

col1_count = collections.Counter([s[0] for s in l])
print(l)
l = sorted(result, key=lambda x:-col1_count[x])
print(l)
```
